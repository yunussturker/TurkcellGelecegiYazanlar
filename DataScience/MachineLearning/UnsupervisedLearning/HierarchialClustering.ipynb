### Hierarchical Clustering
#### Amaç, gözlemleri birbirlerine olan benzerliklerine göre alt kümelere ayırmaktır.
##### Gözlemler daha fazla sayıda alt kümeye ayrılmak istenildiğinde kullanılabilir.
###### K-Means yönteminde sadece belirli sayıda kümeye ayırabiliyoruz.
###### Hiyerarşik kümelemede ise hem belirli sayıda kümeye ayırabiliriz hem de oluşan yeni kümelerin altında da yeni kümelere ayırma işlemi yapabiliriz.

#### Birleştirici Kümeleme
##### Başlangıçta gözlem kadar küme vardır.
###### Veri setinde birbirine en yakın olan iki gözlem bulunur.
###### Bu iki  nokta bir araya geliştirilerek yeni bir gözlem oluşturulur. Yani artık veri seti ilk birleşimdeki işlemlerden oluşmaktadır.
###### Aynı işlemi tekrarlanarak yukarı doğru çıkılır. Yani iki kümenin birleşiminden oluşan bu yeni kümeler aynı şekilde birbirlerine benzerliklerine göre tekrar birleştirilir. Bu işlem tüm gözlemler tek bir kümede toplanana kadar bu işlem tekrar edilir.
##### Birbirine yakın noktalar, uzaklık ölçüleri kullanarak belirlenir.(Öklit uzaklığı, manhattan uzaklığı, korelasyon vs)

#### Bölümleyici Kümeleme
##### Başlangıçta 1 tane küme vardır, o da tüm veri setidir.
###### Tüm gözlemlerin bir arada olduğu küme iki alt kümeye ayrılır.
###### Oluşan yeni kümeler birbirlerine benzemeyen alt kümelere bölünür.
###### Aynı işlem, gözlem sayısı kadar küme elde edinceye kadar tekrar edilir.

#### Kümeleme Yöntemleri ile Karar Ağaçlarının Karşılaştırılması
##### Hiyerarşik yöntemlerde küme sayısına dendrogram sonuçlarına göre kadar verilirken, hiyerarşik olmayan yöntemlerde küme sayısı uygulama yapılmadan önce belirlenir.
##### Hiyerarşik kümeleme yöntemlerinde veri seti gözlemler ya da değişkenler bazında kümeleme işlemine sokulabilirken, hiyerarşik olmayan yöntemlerde sadece gözlemlerin kümelenmesi mümkündür.
##### Karar ağaçlarından farkı; karar ağaçlarında ayırma işlemi hedef değişkene göre yapılırken burada bağımlı değişken olmadığı için gözlemler bağımsız değişkenler üzerinden yapılan uzaklık hesaplarına göre kümelere ayrılır.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("USArrests.csv", index_col=0)
```

```python
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
plt.figure(figsize=(10,5))
plt.title("Hiyerarşik Kümeleme Dendrogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_complete,
          leaf_font_size = 10);
```

```python
plt.figure(figsize=(10,5))
plt.title("Hiyerarşik Kümeleme Dendrogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_average,
          leaf_font_size = 10);
```

```python
plt.figure(figsize=(15,10))
plt.title("Hiyerarşik Kümeleme Dendrogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_complete,
           truncate_mode = "lastp",
           p = 10,
           show_contracted = True,
           leaf_font_size = 10);
```
